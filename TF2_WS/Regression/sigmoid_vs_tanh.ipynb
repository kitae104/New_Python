# sigmoid와 tanh 그리고 relu 함수 비교 

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5, 5, 0.01)
sigmoid_x = tf.math.sigmoid(x)
tanh_x = tf.math.tanh(x)
relu_x = tf.nn.relu(x)
```

#### 그래프 그리기

```python
plt.axhline(0, color='gray')    # 가로선 긋기
plt.axvline(0, color='gray')    # 세로선 긋기
plt.plot(x, sigmoid_x, 'b-', label='sigmoid')
plt.plot(x, tanh_x, 'r--', label='tanh')
plt.plot(x, relu_x, 'g-', label='relu', linewidth=3)
plt.legend()
plt.show()
```

#### 지수 함수 그래프 
* 큰 값을 강조하고 작은 값은 약화하는 효과를 가짐.

```python
import matplotlib.pyplot as plt
import math
import numpy as np
x = np.arange(-2, 2, 0.01)
e_x = math.e ** x

plt.axhline(0, color='gray')
plt.axvline(0, color='gray')
plt.plot(x, x, 'b-', label='y=x')
plt.plot(x, e_x, 'g.', label='y=e^x')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
