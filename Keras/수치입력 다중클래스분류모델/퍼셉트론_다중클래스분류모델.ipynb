# 수치 입력 다중클래스 분류모델(퍼셉트론)

## 1. 데이터셋 준비 

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
import random
```

```python
# 데이터셋 생성
x_train = np.random.random((1000, 12))
y_train = np.random.randint(2, size=(1000, 1))
y_train = to_categorical(y_train, num_classes=10) # 원-핫 인코딩 

x_test = np.random.random((100, 12))
y_test = np.random.randint(2, size=(100, 1))
y_test = to_categorical(y_test, num_classes=10) # 원-핫 인코딩 
```

## 2. 모델 구성하기 

```python
model = Sequential()
model.add(Dense(10, input_dim=12, activation='softmax'))   # 12개 입력 10개 출력  + 10개의 bias 
model.summary()
```

## 3. 모델 학습과정 설정하기

```python
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

## 4. 모델 학습시키기 

```python
hist = model.fit(x_train, y_train, epochs=1000, batch_size=64)
```

## 5. 그래프로 학습과정 살펴보기

```python
%matplotlib inline

fig, loss_ax = plt.subplots()

acc_ax = loss_ax.twinx()

loss_ax.set_ylim([0.0, 1.0])
acc_ax.set_ylim([0.0, 1.0])

loss_ax.plot(hist.history['loss'], 'y', label='train loss')
acc_ax.plot(hist.history['accuracy'], 'b', label='train acc')

loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('accuray')

loss_ax.legend(loc='upper left')
acc_ax.legend(loc='lower left')

plt.show()
```

## 6. 모델 평가하기

```python
loss_and_metrics = model.evaluate(x_test, y_test, batch_size=32)
print('loss_and_metrics : ' + str(loss_and_metrics))
```
