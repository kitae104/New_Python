# 데이터의 적절한 표현

```python
import pandas as pd
```

```python
bank_data = pd.read_csv('data/bank_data_feats_e2.csv', index_col=0)
bank_data.head()
```

```python
bank_data['was_contacted'] = bank_data['pdays'].apply(lambda row : 0 if row == -1 else 1) # 새로운 정보 추가 
```

```python
bank_data.drop('pdays', axis=1, inplace=True)  # 원래 정보 삭제 
```

```python
bank_data[['was_contacted']].head()
```

```python
bank_data.to_csv('data/bank_data_feats_e3.csv')
```
