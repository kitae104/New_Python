# 신경망을 활용한 당뇨병 진단 

```python
import numpy as np
```

```python
data = np.loadtxt("./data/pima-indians-diabetes.csv", delimiter=",")
X = data[:,0:8]
y = data[:,8]
print("Number of Examples in the Dataset = ", X.shape[0])
print("Number of Features for each example = ", X.shape[1]) 
print("Possible Output Classes = ", np.unique(y))
```

```python
seed = 1
np.random.seed(seed)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)  # 70:30 분리
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
print ("Number of examples in training set = ", X_train.shape[0])
print ("Number of examples in test set = ", X_test.shape[0])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(units=8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
hist = model.fit(X_train, y_train, batch_size=5, epochs=300, validation_data=(X_test, y_test))
```

```python
import matplotlib.pyplot as plt 
import matplotlib
%matplotlib inline 
```

```python
plt.rcParams['figure.figsize'] = (10.0, 8.0)
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['lines.color'] = 'r'
plt.rcParams['axes.grid'] = True 
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train loss', 'test loss'], loc='upper right')
plt.show()
```

```python
print("Best Accuracy on training set = ", max(hist.history['accuracy'])*100)
print("Best Accuracy on test set = ", max(hist.history['val_accuracy'])*100)
```

```python
np.random.seed(seed)
model2 = Sequential()
model2.add(Dense(units=16, activation='relu', input_dim=8))
model2.add(Dense(units=8, activation='relu'))
model2.add(Dense(units=1, activation='sigmoid'))
model2.summary()
```

```python
model2.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
hist2 = model2.fit(X_train, y_train, batch_size=5, epochs=350, validation_data=(X_test, y_test))
```

```python
plt.rcParams['figure.figsize'] = (10.0, 8.0)
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['lines.color'] = 'r'
plt.rcParams['axes.grid'] = True 
```

```python
plt.plot(hist2.history['loss'])
plt.plot(hist2.history['val_loss'])
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train loss', 'test loss'], loc='upper right')
plt.show()
```

```python
print("Best Accuracy on training set = ", max(hist2.history['accuracy'])*100)
print("Best Accuracy on test set = ", max(hist2.history['val_accuracy'])*100)
```

```python
np.random.seed(seed)
model3 = Sequential()
model3.add(Dense(units=16, activation='relu', input_dim=8))
model3.add(Dense(units=8, activation='relu'))
model3.add(Dense(units=4, activation='relu'))
model3.add(Dense(units=1, activation='sigmoid'))
model3.summary()
```

```python
hist3 = model3.fit(X_train, y_train, batch_size=5, epochs=400, validation_data=(X_test, y_test))
```
