# 시각화 (그래프 그리기)

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

#### 간단한 꺾은선 그래프 그리기

```python
x = range(20)                       # x 축 범위
y = tf.random.normal([20],0,1)      # y 축 임의의 값 
plt.plot(x,y)                       # 기본 선 그래프
plt.show()
```

####  간단한 점 그래프 그리기

```python
x = range(20)
y = tf.random.normal([20],0,1)
plt.plot(x,y,'bo')               # 파란색의 둥근점 
plt.show()
```

#### 정규분포 그래프를 히스토그램으로 나타내기

```python
random_normal = tf.random.normal([100000],0,1)
plt.hist(random_normal, bins=100)
plt.show()
```
