# Scikit-Learn을 활용한 Iris Clustring

```python
# 1 데이터 가져오기
from sklearn.datasets import load_iris

iris = load_iris()
print(iris)
```

```python
# 2. 데이터 살펴보기 
print(iris.keys())
```

```python
print(iris.data)
```

```python
print(iris.feature_names)
```

```python
print(iris.target)
```

```python
print(iris.target_names)
```

```python
# 3. X / Y 데이터 나누기 
X = iris.data
y = iris.target

print(X.shape)
print(y.shape)
```

```python
# 4. 데이터 시각화 
import matplotlib.pyplot as plt 
plt.scatter(X[:, 1], X[:, 2], c = y)
plt.xlabel('Sepal Length', fontsize=15)
plt.ylabel('Sepal Width', fontsize=15)
plt.show()
```

```python
# 5. Model Train 
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3, n_jobs=1, random_state=23)
km.fit(X)
```

```python
# 6. Centroid 확인 
centers = km.cluster_centers_
print(centers)
```

```python
# 7. 그룹 결과 확인 
new_labels = km.labels_
print(new_labels)
```

```python
# 8. 결과 시각화 
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

#실제 데이터 
axes[0].set_title("Actual", fontsize=15)
axes[0].scatter(X[:, 1], X[:, 2], c=y, s=150)
axes[0].set_xlabel('Sepla length', fontsize=15)
axes[0].set_ylabel('Sepla width', fontsize=15)

# K-Means Clustering 데이터 
axes[1].set_title("KMeans", fontsize=15)
axes[1].scatter(X[:, 1], X[:, 2], c=new_labels, cmap='jet', s=150)
axes[1].set_xlabel('Sepla length', fontsize=15)
axes[1].set_ylabel('Sepla width', fontsize=15)
```
