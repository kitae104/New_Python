# Scikit-Learn을 이용한 기후 변화 예측 모델 

## 1. numpy 형태로 파일 받아오기 

```python
import numpy as np 

year, temp_diff = np.loadtxt('5-year-mean-1882-2014.csv', delimiter=',', usecols=(0, 1), unpack=True)

# 데이터 확인 
print(year[:9])
print(temp_diff[:9])
```

## 2. 데이터 시각화 하기

```python
import matplotlib.pyplot as plt

plt.scatter(year, temp_diff)
plt.title('scatter plot of temp diff vs year')
plt.xlabel('year')
plt.ylabel('time diff')
plt.show()
```

## 3. Regression Model 

```python
from  sklearn.linear_model import LinearRegression

linear_model = LinearRegression(fit_intercept=True)  # y 절편 사용 여부 

year = year.reshape(133, 1)  # 모양 변경 

linear_model.fit(year, temp_diff)
```

## 4 모델 예측

```python
model_predict = linear_model.predict(year)
print(model_predict[0:9])
```

## 5. 모델 시각화 

```python
plt.scatter(year, temp_diff)
plt.plot(year, model_predict)
plt.title('real temp diff vs. model predicted temp diff')
plt.show()
```

## 6. model 확인 

```python
print('y={0} * X + {1}'.format(linear_model.coef_[0], linear_model.intercept_))
```
